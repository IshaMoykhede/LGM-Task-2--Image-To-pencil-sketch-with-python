**Importing the necessary libraries**

```python
from google.colab.patches import cv2_imshow
import cv2
```

**Read The Original Image**

```python
img = cv2.imread('/content/drive/MyDrive/dog.jpg', 1)
cv2_imshow(img)
```

**Converting an image into gray_scale image**

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

**Invert The Gray Scale Image**

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

**Smoothing the Image**

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

**Blur the Inverted Image**

```python
img_blur=cv2.GaussianBlur(img_invert, (111,111),0)
cv2_imshow(img_blur)
```

**Invert the Blurred Image**

```python
invblur_img=cv2.bitwise_not(img_blur)
cv2_imshow(invblur_img)
```

**Obtaining the final sketch**

```python
def dodgeV2(x, y):

    return cv2.divide(x, 255 - y, scale=256)
```

**Final Output as Pencil Sketcht**

```python
final_img = dodgeV2(img_gray, img_smoothing)

cv2_imshow(final_img)

```
